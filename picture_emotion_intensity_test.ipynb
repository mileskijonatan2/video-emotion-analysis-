```python
from deepface import DeepFace
```

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder_path = '/content/drive/MyDrive/emotion_analysis_intensity/happy'
```

```python
import os
```

```python
file_names = (os.listdir(folder_path))
```

```python
cnt = 0
for file_name in file_names:
  if cnt>80:
    break
  else:
    file_path = os.path.join(folder_path, file_name)
    image = cv2.imread(file_path)
    results = DeepFace.analyze(image, actions="emotion", enforce_detection=False)
    emotion_probability = results[0]['emotion']['happy']
    cnt += 1
    plt.figure(figsize=(2, 2))
    plt.imshow(image)
    plt.text(3, -3, f'happiness probability: {emotion_probability}')
    plt.show()
```
